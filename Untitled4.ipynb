```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Membaca citra dalam mode grayscale
img = cv2.imread("/content/download (16).jpg", 0)

if img is None:
    print("Error: Gambar tidak ditemukan. Periksa kembali path filenya.")
else:
    # Kernel Prewitt Horizontal
    kernelx = np.array([[-1, 0, 1],
                        [-1, 0, 1],
                        [-1, 0, 1]], dtype=np.float32)

    # Kernel Prewitt Vertikal
    kernely = np.array([[-1, -1, -1],
                        [0, 0, 0],
                        [1, 1, 1]], dtype=np.float32)

    # Konvolusi
    prewittx = cv2.filter2D(img, -1, kernelx)
    prewitty = cv2.filter2D(img, -1, kernely)

    # Menggabungkan hasil
    prewitt_combined = cv2.add(prewittx, prewitty)

    # Menampilkan hasil di Colab
    print("Prewitt X:")
    cv2_imshow(prewittx)
    print("Prewitt Y:")
    cv2_imshow(prewitty)
    print("Prewitt Combined:")
    cv2_imshow(prewitt_combined)



```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Pastikan ekstensi file benar (biasanya .jpg bukan .img)
img = cv2.imread("/content/download (16).jpg", 0)

if img is None:
    print("Error: File gambar tidak ditemukan! Periksa nama dan ekstensinya.")
else:
    # Menghitung gradient Sobel
    sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
    sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

    # Menghitung magnitudo gradient gabungan
    sobel_combined = np.sqrt(sobelx**2 + sobely**2)

    # Pencegahan error jika sobel_combined.max() adalah 0
    if sobel_combined.max() != 0:
        sobel_combined = np.uint8(sobel_combined / sobel_combined.max() * 255)
    else:
        sobel_combined = np.uint8(sobel_combined)

    # 2. Menampilkan hasil menggunakan Matplotlib (Solusi untuk Colab/Notebook)
    plt.figure(figsize=(12, 8))

    plt.subplot(2, 2, 1); plt.imshow(img, cmap='gray'); plt.title("Original"); plt.axis('off')
    plt.subplot(2, 2, 2); plt.imshow(np.abs(sobelx), cmap='gray'); plt.title("Sobel X"); plt.axis('off')
    plt.subplot(2, 2, 3); plt.imshow(np.abs(sobely), cmap='gray'); plt.title("Sobel Y"); plt.axis('off')
    plt.subplot(2, 2, 4); plt.imshow(sobel_combined, cmap='gray'); plt.title("Sobel Combined"); plt.axis('off')

    plt.tight_layout()
    plt.show()
```
